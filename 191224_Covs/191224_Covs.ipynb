### Example

```python
def gauss_combine(rx, ux, varx, ry, uy, vary, covxy):
    '''Returns uz, varz for Z = rx * X + ry * Y
    
    Examples:
    >>> gauss_combine(1, 5, 10, 0, 123, 456, 789)
    (5, 10)
    >>> gauss_combine(2, 5, 10, 0, 123, 456, 789)
    (10, 40)
    >>> gauss_combine(1, 5, 10, 1, 6, 11, 0)
    (11, 21)
    >>> gauss_combine(1, 5, 10, 1, 5, 10, 0)
    (10, 20)
    >>> gauss_combine(1, 5, 10, 1, 5, 10, 1.0 * sqrt(10) * sqrt(10))
    (10, 40.0)
    >>> gauss_combine(1, 5, 10, 1, 5, 10, 0.5 * sqrt(10) * sqrt(10))
    (10, 30.0)
    ''' 
    uz = rx * ux + ry * uy
    varz = rx**2 * varx + ry**2 * vary + 2 * rx * ry * covxy
    return uz, varz
```

```python
ux = 1.1
varx = .2**2
uy = 1.05
vary = .2**2
covxy = 0
```

```python
rr = linspace(0, 1, 50)
```

```python
figsize(17,4)
uz, varz = gauss_combine(1 - rr, ux, varx, rr, uy, vary, covxy)

subplot(1,3,1)
plot(rr, uz, 'k')
plot(rr[0], uz[0], 'bo', ms=10)
plot(rr[-1], uz[-1], 'ro', ms=10)
xlabel('r'); ylabel('u')

subplot(1,3,2)
plot(rr, varz, 'k')
plot(rr[0], varz[0], 'bo', ms=10)
plot(rr[-1], varz[-1], 'ro', ms=10)
xlabel('r'); ylabel('var')

subplot(1,3,3)
plot(uz, varz, 'k')
plot(uz[0], varz[0], 'bo', ms=10)
plot(uz[-1], varz[-1], 'ro', ms=10)
xlabel('u'); ylabel('var')
```
